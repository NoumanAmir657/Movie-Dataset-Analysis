### Adding Genres list to movie_df

```python
movie_df
```

```python
s = movie_df['Genre'].str.split(',').apply(Series, 1).stack()
s.index = s.index.droplevel(-1)
s.name = 'Genre'
del movie_df['Genre']
movie_df1 = movie_df.join(s)
```

```python
movie_df1
```

### Grouping by Movie's Genres

```python
movie_df_groups = (pd.DataFrame(movie_df1.groupby('Genre').Title.nunique())).sort_values('Title', ascending=False)
```

```python
movie_df_groups
```

```python
movie_df_groups[['Title']].plot.barh(stacked=True, title = 'No. of Movies According to Genre', figsize=(8, 8));
```

## Revenue and Genre

```python
action_df = movie_df1[movie_df1.Genre == 'Action']
adv_df = movie_df1[movie_df1.Genre == 'Adventure']
fantasy_df = movie_df1[movie_df1.Genre == 'Fantasy']
scifi_df = movie_df1[movie_df1.Genre == 'Sci-Fi']
thriller_df = movie_df1[movie_df1.Genre == 'Thriller']
rom_df = movie_df1[movie_df1.Genre == 'Romance']
ani_df = movie_df1[movie_df1.Genre == 'Animation']
comedy_df = movie_df1[movie_df1.Genre == 'Comedy']
fam_df = movie_df1[movie_df1.Genre == 'Family']
musical_df = movie_df1[movie_df1.Genre == 'Musical']
mys_df = movie_df1[movie_df1.Genre == 'Mystery']
western_df = movie_df1[movie_df1.Genre == 'Western']
drama_df = movie_df1[movie_df1.Genre == 'Drama']
hist_df = movie_df1[movie_df1.Genre == 'History']
sport_df = movie_df1[movie_df1.Genre == 'Sport']
crime_df = movie_df1[movie_df1.Genre == 'Crime']
horr_df = movie_df1[movie_df1.Genre == 'Horror']
war_df = movie_df1[movie_df1.Genre == 'War']
bio_df = movie_df1[movie_df1.Genre == 'Biography']
music_df = movie_df1[movie_df1.Genre == 'Music']
```
